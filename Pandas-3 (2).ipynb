```python
import csv
import random
from faker import Faker

# Initialize Faker to generate random data
fake = Faker()

# Define departments and cities
departments = ["HR", "Engineering", "Marketing", "Sales", "Finance", "Operations", "Customer Support"]
cities = ["New York", "Los Angeles", "Chicago", "San Francisco", "Houston", "Miami", "Dallas", "Atlanta", "Seattle", "Boston"]

# Create the CSV file with 100 records
filename = "employees.csv"

fields = ["eid", "ename", "city", "department"]
rows = []

for eid in range(1, 101):
    ename = fake.name()
    city = random.choice(cities)
    department = random.choice(departments)
    rows.append([eid, ename, city, department])

# Write the data to CSV file
with open(filename, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(fields)
    writer.writerows(rows)

print(f"CSV file '{filename}' has been created successfully!")
```

```python
4. DataFrame Operations

df.apply() – Apply a function along an axis (rows or columns).
df.applymap() – Apply a function element-wise for DataFrames.
df.transform() – Apply a function to each group or column independently.
df.map() – Map values to a Series with a function or dictionary.
df.pivot() – Reshape data by specifying columns as index/values.
df.pivot_table() – Pivot the data with aggregation functions.
df.melt() – Unpivot or reshape data.
df.stack() – Stack the DataFrame into a series.
df.unstack() – Unstack (inverse of stack).
df.sort_values() – Sort by one or more columns.
df.sort_index() – Sort by index.
df.reset_index() – Reset the index.
df.set_index() – Set a column as index.
```

```python
df.add() – Add two DataFrames element-wise.
df.subtract() – Subtract two DataFrames element-wise.
df.multiply() – Multiply two DataFrames element-wise.
df.divide() – Divide two DataFrames element-wise.
df.mod() – Modulo (remainder) of two DataFrames element-wise.
df.abs() – Get the absolute value of the DataFrame.
df.round() – Round each element to a given number of decimals.
```
