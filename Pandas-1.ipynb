```python
from openpyxl import load_workbook

def concat_df(dflist):
    finaldf = pd.concat(dflist)
    return finaldf

def get_sheetnames(filename):
    wb = load_workbook(filename)
    return wb.sheetnames
    

def main():
    filename = 'empdata.xlsx'
    sheets = get_sheetnames(filename)
    dfl = []
    for sheet in sheets:
        dfl.append(pd.read_excel(filename,sheet_name=sheet))
    final = concat_df(dfl)
    print(final)
if __name__=='__main__':
    main()
```
