```python
df = pd.DataFrame({"studentname":['suresh','naresh','amit','sonali'],
                   "math":[46,34,75,34],
                   "science":[45,23,76,34],
                   "english":[65,23,75,34]
                  })
```

```python
df
```

```python
df1 = pd.melt(df,id_vars=["studentname"],var_name="Subjects",value_name="Marks")
```

```python
df1
```

```python
final = df1.pivot(index="studentname",columns="Subjects",values="Marks")
```

```python
temp = final.reset_index()
```

```python
temp.drop("Subjects",axis=1)
```

```python
temp.reset_index(inplace=True)
```

```python
temp.to_csv("temp1.csv",index=False)
```
